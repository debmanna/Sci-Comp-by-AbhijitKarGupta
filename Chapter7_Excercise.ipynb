```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = 1/(1+np.exp(-x)) # a Sigmoidal function

z = np.diff(y)/np.diff(x)

plt.subplot(2,1,1)
plt.plot(x,y)

plt.subplot(2,1,2)
plt.plot(x[:-1],z)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,10,1000)
x = 1/(1+np.exp(-2*t + 4))

velocity = np.diff(x)/np.diff(t)
acceleration = np.diff(velocity)/np.diff(t)[:-1]

plt.subplot(1,3,1)
plt.plot(t[:-1],velocity)
plt.title("time vs Velocity")

plt.subplot(1,3,2)
plt.plot(t[:-2],acceleration)
plt.title("time vs Acceleration")

plt.subplot(1,3,3)
plt.plot(t,x)
plt.title("time vs Position")

plt.show()
```

```python
# Question7

import numpy as np

f1 = open("test1.txt",'a')

for i in range(1,7):
    m = 10**i
    print(i,"SampleSPACE_SIZE =",m,"\n",file=f1)
    
    for j in range(1,6):
        arr = np.random.randint(0,2,m) # Head is 1 and Tail is 0
        
        countOf_1 = (arr == 1).sum()
        countOf_0 = m - countOf_1
        print("Experiment",j,file=f1)
        print("The Probability of Getting Head:",countOf_1/m,file=f1)
        print("The Probability of Getting Tail:",countOf_0/m,file=f1)
    
    print("\n",file=f1)
    
f1.close()
```
